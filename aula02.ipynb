Programa que avalia as notas do nosso aplicativo inseriu.

```python
avaliacao = int(input("Digite a sua avaliação 0 - 10 estrelas"))
if avaliacao > 8:
    print("Nota excelente")
elif avaliacao > 6:
    print("Avaliação regular")
elif avaliacao > 4:
    print("Avaliação ruim")
else:
    print("Avaliação péssima")
```

Sitema que irá verificar se o filme avaliado é bom ou não.
- Criar uma variável que irá armazenar o nome do filme
- criar uma variável que irá armazenar a nota do filme (0 - 5)
- Condição: Se possuir 5 estrelas, filme Vencedor do Oscar.
- Se possuir de 3 a 4,99, filme Bom para assistir
- Abaixo disso, filme não indicado

```python
filme = (input("Digite o nome do filme"))
avaliacao_filme = float(input("Digite a sua nota para o filme 0 - 5"))
if avaliacao_filme <= 0 or avalicao_filme >5:
    print("❌ Nota Invalida")
else:
    if avaliacao_filme == 5:
        print(f"filme {filme} Vencedor do Oscar")
    if avaliacao_filme >= 3 and avaliacao_filme < 4.99: 
        print(f"filme {filme} Bom para assistir")
    if avaliacao_filme < 3:
        print(f"filme {filme} Não indicado")

```
