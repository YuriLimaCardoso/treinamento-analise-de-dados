```python
#Armazenar em uma variavel
df = pd.read_csv(r"data\dataset_techshop_vendas.csv",parse_dates=["data_pedido"])
```

```python
df.info()
```

```python
#Exibe as colunas em formato de lista
df.columns.to_list()
```

```python
df.head()
```

```python
df["faturamento"] = df["preco_unitario"] * df["quantidade"]
df.head()
```

```python
df[["preco_unitario","quantidade","faturamento"]]
```

```python
faturamento_total = df["faturamento"].sum()
print(faturamento_total)
```

```python
faturamento_formatado = f"{faturamento_total:,.2f}".replace(",","x").replace(".",",").replace("x",".")
```

```python
print(f"O faturamento total foi de R${faturamento_formatado}")
```

```python
#df.loc['total'] = df.sum(numeric_only=True)
#df.tail()
```

```python
df.loc["total","data_pedido"] = "-"
df.loc["total","categoria"] = "-"
df.loc["total","regiao"] = "-"
df.loc["total","canal_marketing"] = "-"
df.tail()

```

```python
#xibe uma celula especifica
df.loc[799]
```

```python
#Cria um novo arquivo csv a partir do que voce ja tem
df.to_csv("dataset_com_total.csv")
```

```python
grafico = sns.countplot(data=df,x="categoria");
grafico.set_title("Pedidos por categoria");
grafico.set_xlabel("Categorias");
grafico.set_ylabel("Quantidades");
```

```python
grafico = sns.boxplot(data=df,x="categoria",y="preco_unitario");
grafico.set_title("BoxPlot");
grafico.set_ylabel("Preços");
grafico.set_xlabel("Categorias");
```

```python
grafico = sns.histplot(data=df,x="preco_unitario",bins=40,kde=True);
grafico.set_xlabel("Preço unitario");
grafico.set_ylabel("Quantidade");
grafico.set_title("Histograma");
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r"data\dataset_aula02_exercicio_modafina.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df["faturamento"] = df["preco"] * df["quantidade"]
df.head()
```

```python
faturamento_total = df["faturamento"].sum()
print(f"O faturamento total foi de R${faturamento_total}")
```

```python
faturamento_formatado = f"{faturamento_total:,.2f}".replace(",","x").replace(".",",").replace("x",".")
print(faturamento_formatado)
```

```python
df["categoria_produto"].value_counts().sort_values(ascending=False)
```

```python
#df.groupby["categoria_produto"]("faturamento")
```

```python
df.groupby("categoria_produto")["faturamento"].sum().sort_values(ascending=False)
```

```python
grafico = sns.histplot(data=df,x="categoria_produto");
grafico.set_title ("Pedidos por categoria");
grafico.set_xlabel("Produto");
grafico.set_ylabel("Quantidade");
```

```python
grafico = sns.boxplot(data=df, x="categoria_produto",y="preco");
grafico.set_title("Preço de cada produto");
grafico.set_xlabel("Produto");
grafico.set_ylabel("Preço");
```

```python
grafico = sns.histplot(data=df, x="preco", bins=20, kde=True);
grafico.bar_label(grafico.containers[0])
grafico.set_title("Distribuição dos preços dos produtos");
grafico.set_xlabel("Preço");
grafico.set_ylabel("Quantidade");
```
