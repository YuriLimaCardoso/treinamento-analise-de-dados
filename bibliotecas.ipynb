```python
#Importação de bibliotecas
import numpy as np #np será o apelido para numpy
```

```python
temperaturas = np.array([30,40,50,60,70,80,90,100])
print(temperaturas)
print(type(temperaturas))
```

```python
#Media artimética
media = np.mean(temperaturas)
print(media)
```

```python
#Retorna o maior valor
maior_valor = np.max(temperaturas)
print(maior_valor)
```

```python
menor_valor = np.min(temperaturas)
print(menor_valor)
```

```python
#Salarios x1000
salarios = np.array([2,3,4,5,1,500])
media_salarios = np.mean (salarios)
print(f"{media_salarios:.2f}")
```

```python
mediana = np.median(salarios)
print(mediana)
```

Desafio Final
- Criar um array de habitantes de 5 idade do estado de sp
- Maior população
- Menor população
- Media da populacao
- Mediana

```python
#SP, GRU, CAMPINAS, SÃO BERNARDO DO CAMPO, SANTO ANDRE
import numpy as np

# Nomes das cidades
cidades = np.array([
    "São Paulo",
    "Guarulhos",
    "Campinas",
    "São Bernardo do Campo",
    "Santo André"
])

# População das cidades
habitantes_cidades_sp = np.array([
    11904961,
    1349100,
    1187974,
    841154,
    782448
])


for cidade, habitantes in zip(cidades, habitantes_cidades_sp):
    print(f"{cidade}: {habitantes:,} habitantes")
    
habitantes_cidades_sp = np.array([11904961,1349100,1187974,841154,782448])
print(habitantes_cidades_sp)
```

```python
maior_valor_habitantes = np.max(habitantes_cidades_sp)
print(maior_valor_habitantes)
```

```python
menor_valor_habitantes = np.min(habitantes_cidades_sp)
print(menor_valor_habitantes)
```

```python
media_valor_habitantes = np.mean(habitantes_cidades_sp)
print(media_valor_habitantes)
```

```python
mediana_habitantes = np.median(habitantes_cidades_sp)
print(mediana_habitantes)
```
